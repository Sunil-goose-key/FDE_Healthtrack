# HealthTrack Profiling

```python
import pandas as pd
from ydata_profiling import ProfileReport

df = pd.read_csv('healthtrack_raw.csv')
profile = ProfileReport(df, title='HealthTrack Data Audit', explorative=True)
profile.to_file('healthtrack_profile.html')
df.head()
```

## Audit Findings

- Rows missing `patient_id`: **25** (8.33% of 300 records).
- Rows missing `readmission_30d`: **30**. A missing outcome value blocks supervised model training because the model needs known outcomes to learn what risk patterns lead to readmission.
- Rows with `total_bill_usd == -9999`: **15**. This value likely represents an EHR placeholder for unavailable billing data rather than a true patient bill.
- Department with highest readmission proportion: **Oncology**.

The most critical issue is missing 30-day readmission outcomes, because it prevents us from training and validating a reliable model to identify patients needing early intervention.
